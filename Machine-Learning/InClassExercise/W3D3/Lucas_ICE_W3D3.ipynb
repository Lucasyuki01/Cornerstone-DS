```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
df = pd.read_csv('new_retail_data.csv')
```

```python
num_cols = ["Age", "Total_Purchases", "Amount", "Total_Amount", "Ratings"]
scaling_df = df[num_cols].copy()
```

```python
standard = StandardScaler()
robust = RobustScaler()
minmax = MinMaxScaler()
```

```python
# StandardScaler: for normally distributed data (Age)
scaling_df["Age_scaled"] = standard.fit_transform(scaling_df[["Age"]])
```

```python
# RobustScaler: for features with outliers (Total_Purchases, Amount, Total_Amount)
scaling_df[["Total_Purchases_scaled", "Amount_scaled", "Total_Amount_scaled"]] = robust.fit_transform(
    scaling_df[["Total_Purchases", "Amount", "Total_Amount"]]
)
```

```python
# MinMaxScaler: for bounded scale [1â€“5] (Ratings)
scaling_df["Ratings_scaled"] = minmax.fit_transform(scaling_df[["Ratings"]])
```
