```python
import math
import numpy as np
```

```python
string = [10, 20, 60, 70, 80, 90, 1000]
```

```python
media = sum(string)/len(string)
```

```python
soma = 0
for i in string:
    soma += i

n = len(string)

x_line = soma/n
```

```python
lista_diferenca = []
for x in string:
    lista_diferenca.append(x - x_line)

print(lista_diferenca)
```

```python
lista_difereca_quadrados = []
for i in lista_diferenca:
    lista_difereca_quadrados.append(pow(i, 2))
    
print(lista_difereca_quadrados)
```

```python
soma_quadrados = 0
for j in lista_difereca_quadrados:
    soma_quadrados += j

print(soma_quadrados)
```

```python
print('Para população:')
variancia = soma_quadrados/n
print(f'variancia = {variancia}')

desvio_padrao = math.sqrt(variancia)
print(f'desvio padrão: {desvio_padrao}')
```

```python
print('Para Amostra:')
variancia2 = soma_quadrados/(n-1)
print(f'variancia = {variancia2}')

desvio_padrao2 = math.sqrt(variancia2)
print(f'desvio padrão: {desvio_padrao2}')
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer, StandardScaler, RobustScaler
import numpy as np
```

```python
x_min = min(string)
x_max = max(string)
```

### Min-Max

```python
#jeito 1

scaleds_minmax = []
for x in string:
    x_line2 = ((x - x_min)/(x_max - x_min))
    scaleds_minmax.append(x_line2)

print(scaleds_minmax[::-1])
```

```python
#Jeito 2

# Redimensionando a lista para uma estrutura 2d
string_minmax = np.array(string).reshape(-1, 1)

scaler_minmax = MinMaxScaler()
X_scaled = scaler_minmax.fit_transform(string_minmax)
print(X_scaled[::-1])
```

### Normalization

```python
#jeito 1

# norma L2
l2_norm = math.sqrt(sum(x*x for x in string))

# normalização
normalized_l2 = [x / l2_norm for x in string]

print(normalized_l2)
```

```python
#jeito 2
# Redimensionando a lista para uma estrutura 2d
string_normalizer = np.array(string).reshape(1, -1)

scaler_normalizer = Normalizer(norm='l2')
X_scaled_normalized = scaler_normalizer.fit_transform(string_normalizer)
print (X_scaled_normalized)
```

### Standard Scaler

```python
#jeito 1

standard_scaled_manual = [(x - media) / desvio_padrao for x in string]
print("Manual:", standard_scaled_manual)
```

```python
#jeito 2
string_standardscaler = np.array(string).reshape(-1, 1)

scaler = StandardScaler()
standard_scaled_sklearn = scaler.fit_transform(string_standardscaler)

print("Sklearn:\n", standard_scaled_sklearn.flatten())
```

### Robust Scaler

```python
#jeito 1
# 1. Convertendo pra numpy
arr = np.array(string)

# 2. Mediana
median = np.median(arr)

# 3. Q1 e Q3
q1 = np.percentile(arr, 25)
q3 = np.percentile(arr, 75)

# 4. IQR
iqr = q3 - q1

# 5. Robust Scaler manual
robust_scaled_manual = [(x - median) / iqr for x in arr]

print("Median:", median)
print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)
print("Manual:", robust_scaled_manual)
```

```python
#jeito 2
string_robustscaler = np.array(string).reshape(-1, 1)

scaler = RobustScaler()
robust_scaled_sklearn = scaler.fit_transform(string_robustscaler)

print("Sklearn:", robust_scaled_sklearn.flatten())
```
